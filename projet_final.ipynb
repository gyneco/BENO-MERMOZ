## Étape 1 : La Classe Étudiant (Jour 5 - POO)
Nous allons créer une classe `Etudiant` qui servira de modèle pour créer nos étudiants. Un étudiant a un nom, un âge et une note.

**Instructions :**
1. Complétez la méthode `__init__` pour assigner les paramètres aux attributs de l'objet (utilisez `self`).
2. Complétez la méthode `afficher_infos` pour retourner une chaîne de caractères formatée (ex: "Nom: Alice, Âge: 20, Note: 15").

```python
class Etudiant:
    def __init__(self, nom, age, note):
        self.nom = nom
        self.age = age
        self.note = note


    def afficher_infos(self):
        return f"Nom: {self.nom}, Âge: {self.age}, Note: {self.note}"
```

## Étape 2 : La Classe Ecole (Jours 1, 3, 4, 5)
Nous allons créer une classe `Ecole` qui va gérer une collection d'étudiants. Nous utiliserons un **dictionnaire** pour stocker les étudiants, où la clé sera le nom de l'étudiant.

**Instructions :**
1. Dans `__init__`, initialisez un dictionnaire vide appelé `self.etudiants`.
2. Dans `ajouter_etudiant`, créez un objet `Etudiant` et ajoutez-le au dictionnaire `self.etudiants` avec le `nom` comme clé.
3. Dans `lister_etudiants`, utilisez une boucle (`for`) pour parcourir et afficher les infos de chaque étudiant du dictionnaire.
4. Dans `moyenne_notes`, parcourez le dictionnaire, additionnez les notes et calculez la moyenne.

```python
class Ecole:
    def __init__(self):
        self.etudiants = {}
        pass

    def ajouter_etudiant(self, nom, age, note):
        etudiant = Etudiant(nom, age, note)
        self.etudiants[nom] = etudiant

        pass

    def lister_etudiants(self):
        print("--- Liste des étudiants ---")
        for etudiant in self.etudiants.values():
                print(etudiant.afficher_infos())
        print("---------------------------")

    def moyenne_notes(self):
        if len(self.etudiants) == 0:
            return 0

        total = 0
        for etudiant in self.etudiants.values():
            total += etudiant.note

        return total / len(self.etudiants)
        return 0
```

## Étape 3 : Le Menu Interactif (Jours 2 et 4)
Maintenant, créons la fonction principale qui fera tourner l'application. Elle affichera un menu et réagira aux choix de l'utilisateur.

**Instructions :**
1. Complétez les conditions `if/elif` pour appeler les bonnes méthodes de la classe `Ecole`.
2. Pour ajouter un étudiant, demandez son nom, son âge et sa note avec `input()`. Pensez à convertir l'âge et la note en entiers (`int()`).
3. Utilisez un bloc `try/except` pour gérer l'erreur si l'utilisateur tape du texte au lieu d'un nombre pour l'âge ou la note.
4. La boucle `while` doit s'arrêter si l'utilisateur tape '4'.

```python
def lancer_application():
    mon_ecole = Ecole()
    print("Bienvenue dans le système de gestion de l'école !")

    while True:
        print("\n1. Ajouter un étudiant")
        print("2. Lister les étudiants")
        print("3. Afficher la moyenne des notes")
        print("4. Quitter")

        choix = input("Votre choix : ")

        if choix == '1':
            nom = input("Entrez le nom de l'étudiant : ")
            try:
                age = int(input("Entrez l'âge de l'étudiant : "))
                note = int(input("Entrez la note de l'étudiant : "))

                mon_ecole.ajouter_etudiant(nom, age, note)

                print(f"{nom} a été ajouté avec succès !")

            except ValueError:
               print("Erreur : l'âge et la note doivent être des nombres.")

        elif choix == '2':
            mon_ecole.lister_etudiants()
            pass

        elif choix == '3':
            moyenne = mon_ecole.moyenne_notes()
            print(f"La moyenne des notes est : {moyenne}")
            pass

        elif choix == '4':
            print("Au revoir !")
            pass

        else:
            print("Choix invalide. Veuillez entrer un nombre entre 1 et 4.")
```
